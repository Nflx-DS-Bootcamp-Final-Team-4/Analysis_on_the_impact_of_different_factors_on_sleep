```python
import pandas as pd
```

```python
# Get the data

all_activity_df_list = []
all_sleep_df_list = []

for num in range(1,23):
    activity_csv_file_path = f'../../Resources/DataPaper/user_{num}/Activity.csv'
    sleep_csv_file_path = f'../../Resources/DataPaper/user_{num}/sleep.csv'
    
    activity_df = pd.read_csv(activity_csv_file_path)
    sleep_df = pd.read_csv(sleep_csv_file_path)
    
    activity_df['User'] = num
    sleep_df['User'] = num
    
    activity_df.set_index(['User'])
    sleep_df.set_index(['User'])

    all_activity_df_list.append(activity_df)
    all_sleep_df_list.append(sleep_df)

all_activity_df = pd.concat(all_activity_df_list)
all_sleep_df = pd.concat(all_sleep_df_list)
```
